2) Compare label distributions across splits (train/val/test)
A) Fix the Violenc column name first

Your split CSVs use Violenc but full uses Violence.

```python
# Fix column name in train/val/test if needed
for df in [df_train, df_val, df_test]:
    if "Violenc" in df.columns and "Violence" not in df.columns:
        df.rename(columns={"Violenc": "Violence"}, inplace=True)

labels = ["Violence", "Hate", "Vulgar", "HateSpeech"]
```

B) Print distribution (counts and percentages) per split
Your split CSVs use Violenc but full uses Violence.

```python
import pandas as pd

def label_stats(df, split_name):
    out = pd.DataFrame({
        "count_1": df[labels].sum(),
        "pct_1": (df[labels].mean() * 100).round(2),
    })
    out["split"] = split_name
    out["n_rows"] = len(df)
    return out.reset_index(names="label")

stats = pd.concat([
    label_stats(df_train, "train"),
    label_stats(df_val, "val"),
    label_stats(df_test, "test"),
], ignore_index=True)

stats_pivot = stats.pivot(index="label", columns="split", values=["count_1", "pct_1"])
stats_pivot
```

C) Check how many labels each tweet has (multi-label intensity)
This is super useful to see if one split has “harder” samples.

```python
def multi_label_intensity(df, split_name):
    k = df[labels].sum(axis=1)  # how many labels = 1 in each row
    return pd.Series({
        "split": split_name,
        "n_rows": len(df),
        "avg_labels_per_tweet": round(k.mean(), 3),
        "pct_any_label": round((k.gt(0).mean() * 100), 2),
        "pct_2plus_labels": round((k.ge(2).mean() * 100), 2),
        "pct_3plus_labels": round((k.ge(3).mean() * 100), 2),
    })

pd.DataFrame([
    multi_label_intensity(df_train, "train"),
    multi_label_intensity(df_val, "val"),
    multi_label_intensity(df_test, "test"),
])
```

D) Verify whether HateSpeech is derived from other labels
Example hypothesis: HateSpeech == (Violence OR Hate OR Vulgar).

```python
def check_hatespeech_rule(df, split_name):
    derived = (df[["Violence","Hate","Vulgar"]].max(axis=1)).astype(int)
    mismatch = (derived != df["HateSpeech"]).sum()
    return split_name, mismatch, len(df)

for name, df in [("train", df_train), ("val", df_val), ("test", df_test)]:
    s, mism, n = check_hatespeech_rule(df, name)
    print(f"{s}: HateSpeech mismatches OR(Violence,Hate,Vulgar) in {mism}/{n}")
```
